```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
import numpy as np
```

```python
d=pd.read_csv('path to StudentsPerformance.csv')
```

```python
sb.catplot(data=d,kind='violin')
```

```python
d.plot.scatter(x='math score',y='reading score',c='red')
d.plot.scatter(x='math score',y='writing score',c='DarkBlue')
```

```python
mathscore=np.array(d.T.iloc[5],dtype=float)
```

```python
x=np.array(d.T[6:8],dtype=float)
```

```python
x=x.reshape(1000,2)
mathscore=mathscore.reshape(1000,1)
```

```python
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import random


fig = pyplot.figure()
ax = Axes3D(fig)

sequence_containing_x_vals = x.T[0]
sequence_containing_y_vals = x.T[1]
sequence_containing_z_vals =mathscore

ax.scatter(x.T[0],x.T[1], mathscore)
ax.set_xlabel('reading')
ax.set_ylabel('writing')
ax.set_zlabel('math score')
pyplot.show()
```

```python
svmreg=svm.SVC(gamma='scale')
svmreg.fit(x,mathscore)
```

```python
result=svmreg.predict(x)
```

```python
fig = pyplot.figure()
ax = Axes3D(fig)
graph=ax.scatter(x.T[1],result)

graph=ax.scatter(x.T[0],result)

ax.set_xlabel('reading')
ax.set_ylabel('writing')
ax.set_zlabel('math score')
```
