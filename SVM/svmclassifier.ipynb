```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d=pd.read_csv('/Users/manthanmkulakarni/PycharmProjects/dataset/StudentsPerformance.csv')
```

```python
y=np.array(d['gender'])
x1=np.array(d['math score'],dtype=int)
x2=np.array(d['reading score'],dtype=int)
x3=np.array(d['writing score'],dtype=int)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure()
fig=Axes3D(fig)

fig.plot(x1,x2,x3,'.')
fig.set_xlabel('Math score')
fig.set_ylabel('Reading score')
fig.set_zlabel('Writing score')
plt.show()
```

```python
temp1=[]
temp2=[]
for i in range(y.shape[0]):
    if(y[i]=='female'):
        temp1.append([x1[i],x2[i],x3[i]])
    else:
        temp2.append([x1[i],x2[i],x3[i]])
temp1=np.array(temp1)
temp2=np.array(temp2)
```

```python
fig=plt.figure()
fig=Axes3D(fig)


fig.plot(temp1.T[0],temp1.T[1],temp1.T[2],'.')
fig.plot(temp2.T[0],temp2.T[1],temp2.T[2],'.')
fig.set_xlabel('Math score')
fig.set_ylabel('Reading score')
fig.set_zlabel('Writing score')
plt.show()
```
